# Inverted Grayscale Image

```python
from PIL import Image, ImageChops
from matplotlib import pyplot as plt

# Open the color image
color_image = Image.open('/kaggle/input/msdhoni/wallpaperflare.com_wallpaper-2.jpg')

# Convert the color image to grayscale
gray_image = color_image.convert('L')

# Invert the grayscale image
inverted_image = ImageChops.invert(gray_image)

# Display the inverted grayscale image with title
plt.imshow(inverted_image, cmap='gray')
plt.title('M.S.Dhoni')
plt.show()
```

# Inverted and Blurred

```python
from PIL import Image, ImageChops, ImageFilter
from matplotlib import pyplot as plt

# Open the color image
color_image = Image.open('/kaggle/input/msdhoni/wallpaperflare.com_wallpaper-2.jpg')

# Convert the color image to grayscale
gray_image = color_image.convert('L')

# Invert the grayscale image
inverted_image = ImageChops.invert(gray_image)

# Applying Blurring to the inverted image using 'GaussianBlur' ImageFilter
inverted_image_blurred = inverted_image.filter(ImageFilter.GaussianBlur(radius=5))

# Display the inverted and blurred grayscale image with title
plt.imshow(inverted_image_blurred, cmap='gray')
plt.title('M.S.Dhoni')
plt.show()
```

# Pencil Sketch Image

```python
import cv2
from PIL import Image
from matplotlib import pyplot as plt

# Open the color image
color_image = Image.open('/kaggle/input/msdhoni/wallpaperflare.com_wallpaper-2.jpg')

# Convert the color image to grayscale
gray_image = color_image.convert('L')

# Apply Gaussian blur to the grayscale image
blurred_image = cv2.GaussianBlur(np.array(gray_image), (21, 21), 0)

# Invert the blurred image
inverted_blurred = 255 - blurred_image

# Create a pencil sketch by dividing the grayscale image by the inverted and blurred image
pencil_sketch = cv2.divide(np.array(gray_image), inverted_blurred, scale=256.0)

# Display the pencil sketch with a title
plt.title("M.S.Dhoni")
plt.imshow(pencil_sketch, cmap='gray')  # Specify 'gray' colormap for grayscale images
plt.show()
```

```python
import cv2
from PIL import Image
from matplotlib import pyplot as plt

# Open the color image
color_image = Image.open('/kaggle/input/msdhoni/wallpaperflare.com_wallpaper-2.jpg')

# Convert the color image to grayscale
gray_image = color_image.convert('L')

# Invert the grayscale image
inverted_image = 255 - np.array(gray_image)

# Apply Gaussian blur to the inverted image
blurred_image = cv2.GaussianBlur(inverted_image, (21,21), 0)

# Create a sketch by dividing the grayscale image by the inverted and blurred image
rgb_sketch = cv2.divide(np.array(gray_image), blurred_image, scale=256.0)

# Plot the original image and its corresponding sketch side by side
plt.figure(figsize=(14, 8))

# Original image
plt.subplot(1, 2, 1)
plt.title('Original image', size=18)
plt.imshow(color_image)
plt.axis('off')

# Sketch
plt.subplot(1, 2, 2)
plt.title('Sketch', size=18)
plt.imshow(rgb_sketch, cmap='gray')  # Specify 'gray' colormap for grayscale images
plt.axis('off')

plt.show()
```
